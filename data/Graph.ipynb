```python
import networkx as nx
import matplotlib.pyplot as plt
import sys
import pandas as pd
```

```python
G = nx.DiGraph() 
G.add_edges_from([(1, 1), (1, 7), (2, 1), (2, 2), (2, 3),  
                  (2, 6), (3, 5), (4, 3), (5, 4), (5, 8), 
                  (5, 9), (6, 4), (7, 2), (7, 6), (8, 7)]) 
  
plt.figure(figsize =(10, 10)) 


node_color = [G.degree(v) for v in G] 
# node colour is a list of degrees of nodes 
  
# node_size = [0.0005 * nx.get_node_attributes(G, 'population')[v] for v in G] 
node_size = [1000*G.degree(v) for v in G]
# size of node is a list of population of cities 
  
# edge_width = [0.0015 * G[u][v]['weight'] for u, v in G.edges()] 
edge_width = [3 for u, v in G.edges()]

# colors=['r','b','y','a']
# width of edge is a list of weight of edges 
  
nx.draw_networkx(G, node_size = node_size,  
                 node_color='#A0CBE2', alpha = 0.7, 
                 with_labels = True, width =1, 
                 edge_color = '0.01', cmap = plt.cm.Blues, 
                 font_size=20, font_weight='bold') 
  
plt.axis('off') 
plt.tight_layout(); 

# plt.text(0.95, 0.9, "Number of Different Edges: "+str(G.number_of_edges()),
#                     verticalalignment='bottom', horizontalalignment='right',
#                     color='green', fontsize=15)
plt.show()

tempDF = pd.DataFrame([dict(G.in_degree()), dict(G.out_degree),
                          nx.algorithms.centrality.degree_alg.degree_centrality(G),
                          nx.algorithms.centrality.closeness.closeness_centrality(G),
                          nx.algorithms.centrality.betweenness_centrality(G)])
df = tempDF.T
       
df.columns = df.columns.map(str)
df.columns = ['a', 'b', 'c','d','e']
df['Times'] = [2] * int(G.number_of_nodes())
df.rename(columns={0: "a", "1": "c", "2":"d", "3":"z", "4":"y"})
print(df)
print(str(df.min(axis=0)['a']))
# # getting different graph attributes 
# print("Total number of nodes: ", int(G.number_of_nodes())) 
# print("Total number of edges: ", int(G.number_of_edges())) 
# print("In-degree for all nodes: ", dict(G.in_degree())) 
# print("Out degree for all nodes: ", dict(G.out_degree)) 
# print("Degree Centrality: ", nx.algorithms.centrality.degree_alg.degree_centrality(G))
# print("Closeness Centrality: ", nx.algorithms.centrality.closeness.closeness_centrality(G))
# print("Betweeness Centrality: ", nx.algorithms.centrality.betweenness_centrality(G))
```
